```python
import pandas as pd
from datetime import datetime
```

```python
coaches = pd.read_csv('edited/coaches.csv')
coaches = coaches.set_index(['tmID', 'year'])

coaches.rename(columns={'stint': 'coachStint'}, inplace=True)
```

```python
teams = pd.read_csv('edited/teams.csv')
teams = teams.set_index(['tmID', 'year'])
```

```python
teamsCoachesComb = coaches.join(teams, how='inner')
teamsCoachesComb
```

```python
players = pd.read_csv('edited/players_teams.csv')
players.set_index(['tmID', 'year'], inplace=True)
```

```python
allComb = teamsCoachesComb.join(players, how='inner')
allComb
```

```python
allComb.to_csv('edited/all.csv')
```
