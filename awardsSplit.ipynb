```python
import pandas as pd
```

```python
awards = pd.read_csv('dataset_edited/awards_players_abrev.csv')
awards
```

```python
coach_awards = awards[awards['award'] == 'COTY']
coach_awards.reset_index()
coach_awards
```

```python
# Create a DataFrame with rows where 'award' is not 'COTY'
player_awards = awards[awards['award'] != 'COTY']
player_awards.reset_index()
player_awards
```

```python
coach_awards.to_csv('dataset_edited/coach_awards.csv', index=False)
player_awards.to_csv('dataset_edited/players_awards.csv', index=False)
```
